```python
import csv
import pandas as pd
import numpy as np

data=pd.read_csv('../public_data/data/사업장성립소멸현황.csv',index_col=0,encoding='utf-8-sig')
print(data.columns)
data.head()
```

```python
data=data.transpose()
data
```

```python
data.drop('총계',axis=1,inplace=True)
for i in range(1,8):
    data[data.columns[i]]=data[data.columns[i]].apply(lambda x : x.replace(',',''))
for i in range(1,8):
    data=data.astype({data.columns[i]:'int64'})

# data["서울특별시"]=data["서울특별시"].apply(lambda x : x.replace(',',''))
# data
```

```python
data.columns[1]
```

```python
year_2018=pd.DataFrame(data.iloc[0:24])
year_2018.head()
year_2018.tail()
```

```python
year_2019=pd.DataFrame(data.iloc[24:48])
year_2019.head()
year_2019.tail()
```

```python
year_2020=pd.DataFrame(data.iloc[48:72])
year_2020.head()
year_2020.tail()
```

```python
year_2021=pd.DataFrame(data.iloc[72:96])
year_2021.head()
year_2021.tail()
```

```python
year_2022=pd.DataFrame(data.iloc[96:120])
year_2022.head()
year_2022.tail()
```

```python
year_2018.set_index('(사업장)시도',inplace=True)
```

```python
year_2018_group=year_2018.groupby('(사업장)시도').sum()
year_2018_group=year_2018_group.rename({'소멸사업장수(월)':'2018',
                                  '신규성립사업장수(월)':'2018'})
new_2018=pd.DataFrame(year_2018_group.iloc[0])
drop_2018=pd.DataFrame(year_2018_group.iloc[1])
```

```python
year_2019_group=year_2019.groupby('(사업장)시도').sum()
year_2019_group=year_2019_group.rename({'소멸사업장수(월)':'2019',
                                  '신규성립사업장수(월)':'2019'})
new_2019=pd.DataFrame(year_2019_group.iloc[0])
drop_2019=pd.DataFrame(year_2019_group.iloc[1])
```

```python
year_2020_group=year_2020.groupby('(사업장)시도').sum()
year_2020_group=year_2020_group.rename({'소멸사업장수(월)':'2020',
                                  '신규성립사업장수(월)':'2020'})
new_2020=pd.DataFrame(year_2020_group.iloc[0])
drop_2020=pd.DataFrame(year_2020_group.iloc[1])
```

```python
year_2021_group=year_2021.groupby('(사업장)시도').sum()
year_2021_group=year_2021_group.rename({'소멸사업장수(월)':'2021',
                                  '신규성립사업장수(월)':'2021'})
new_2021=pd.DataFrame(year_2021_group.iloc[0])
drop_2021=pd.DataFrame(year_2021_group.iloc[1])
```

```python
year_2022_group=year_2022.groupby('(사업장)시도').sum()
year_2022_group=year_2022_group.rename({'소멸사업장수(월)':'2022',
                                  '신규성립사업장수(월)':'2022'})
new_2022=pd.DataFrame(year_2022_group.iloc[0])
drop_2022=pd.DataFrame(year_2022_group.iloc[1])
```

```python
new_enter=pd.concat([new_2018,new_2019,new_2020,new_2021,new_2022],axis=1)
new_enter
```

```python
drop_enter=pd.concat([drop_2018,drop_2019,drop_2020,drop_2021,drop_2022],axis=1)
drop_enter
```

```python
import matplotlib.pyplot as plt
import platform
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
fig=plt.figure(figsize=(20,6))
color= ['cornflowerblue', 'tomato']
plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.35, hspace=0.35)

for i in range(1,8):
    ax1 = fig.add_subplot(2,4,i)
    ax1.plot(new_enter.columns,new_enter.iloc[i-1] ,linestyle = '--',marker = 'o',color=color[0]) # 신규 생성(blue)
    ax1.plot(drop_enter.columns,drop_enter.iloc[i-1] , linestyle = '--',marker = 's',color=color[1]) # 소멸(orange)
    plt.title(new_enter.index[i-1]+"신규 및 소멸 사업장 수 비교")
    plt.xlabel('연도')
    plt.ylabel('시도')
```

```python
import csv
import pandas as pd
import numpy as np

rate=pd.read_csv('../public_data/data/소비자물가지수.csv',index_col=0,encoding='euc-kr')
rate=pd.DataFrame(rate.iloc[10:])
rate
```

```python
job_dp=pd.DataFrame(data[['(지역별)시도','총계']])
job_dp
```

```python
job_dp['총계']=job_dp['총계'].apply(lambda x : x.replace(',',''))
job_dp=job_dp.astype({'총계':'int64'})
job_dp
```

```python
year_2018=pd.DataFrame(job_dp.iloc[0:24])
year_2019=pd.DataFrame(job_dp.iloc[24:48])
year_2020=pd.DataFrame(job_dp.iloc[48:72])
year_2021=pd.DataFrame(job_dp.iloc[72:96])
year_2022=pd.DataFrame(job_dp.iloc[96:120])
```

```python
year_2018_group=year_2018.groupby('(지역별)시도').sum()
year_2019_group=year_2019.groupby('(지역별)시도').sum()
year_2020_group=year_2020.groupby('(지역별)시도').sum()
year_2021_group=year_2021.groupby('(지역별)시도').sum()
year_2022_group=year_2022.groupby('(지역별)시도').sum()
```

```python
job_group=pd.concat([year_2018_group,year_2019_group,year_2020_group,year_2021_group,year_2022_group],axis=1)
job_group
```

```python
job_group=job_group.rename(index={'구인인원(월)':'구인인원','구직건수(월)':'구직건수','취업건수(월)':'취업건수'})
job_group.columns = ['2018','2019','2020','2021','2022']
job_group
```

```python
import matplotlib.pyplot as plt
import numpy as np
import platform

year= ['2018','2019','2020','2021','2022']
fig,ax = plt.subplots(figsize=(12,6))
bar_width = 0.25

index = np.arange(5)

b1 = plt.bar(index,job_group.iloc[0],bar_width,alpha=0.4,color='blue',label='구인인원')
b1 = plt.bar(index+bar_width,job_group.iloc[1],bar_width,alpha=0.4,color='green',label='구직건수')
plt.plot(index,(rate['전국']-min(rate['전국'])+5)*100000,linestyle = '--',marker = 'o',color='red')

plt.xticks(np.arange(bar_width,5 + bar_width ,1),year)

plt.xlabel('연도',size = 13)
plt.ylabel('인원',size = 13)
plt.legend()
plt.show()
```

```python
job_count=job_group.loc['취업건수']
# job_group.loc['구직건수대비취업건수비율']=job_count.div(job_group.loc['구직건수'],axis=0)*100
# job_group
ratio=round(job_count.div(job_group.loc['구직건수'])*100,1)
print(type(ratio))

```

```python
ratio_df=pd.DataFrame(ratio)
ratio_df.columns = ['구직건수 대비 취업건수']
ratio_df
```

```python
import matplotlib.pyplot as plt
import platform

# 판다스 제공 그래프
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
    
mc_ratio = ratio_df[['구직건수 대비 취업건수']]
mc_ratio.plot(kind='bar', figsize = (10,6),color='lightblue',alpha=0.7)
plt.xticks(rotation=0)
plt.twinx()
plt.plot(index,(rate['전국']),linestyle = '--',marker = 'o',color='gray')


```
